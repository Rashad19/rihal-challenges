### Removing low correlation columns

```python
df_train.drop('engine_fuel', inplace=True, axis = 1)
df_train.drop('type_of_drive', inplace=True, axis = 1)
df_train.drop('number_of_photos', inplace=True, axis = 1)
df_train.drop('color', inplace=True, axis = 1)
df_train.drop('body_type', inplace=True, axis = 1)

df_test.drop('engine_fuel', inplace=True, axis = 1)
df_test.drop('type_of_drive', inplace=True, axis = 1)
df_test.drop('number_of_photos', inplace=True, axis = 1)
df_test.drop('color', inplace=True, axis = 1)
df_test.drop('body_type', inplace=True, axis = 1)
```

### Converting categorical variables to Numeric variables

```python
# first we encode the categorical independent variables
# that is, has_warranty, is_exchangeable, transmission, colour, manufacturer_name etc...
# we do this by creating dummy variables for each value and concatenating 
# them to the current data

# defining categorical variables
categorical_columns = ['manufacturer_name', 'transmission', 'engine_type', 
                       'has_warranty', 'is_exchangeable',
                       'ownership']

# generating dummy variables from the categorical variables (TRAINING SET)
df_encode_train = pd.get_dummies(data = df_train, prefix = 'dummy', prefix_sep='_',
               columns = categorical_columns,
               drop_first = True,
               dtype='int8')

# # removing duplicate columns
# df_encode_train = df_encode_train.loc[:,~df_encode_train.columns.duplicated()].copy()

# generating dummy variables from the categorical variables (TEST SET)
df_encode_test = pd.get_dummies(data = df_test, prefix = 'dummy', prefix_sep='_',
               columns = categorical_columns,
               drop_first = True,
               dtype='int8')
```

```python
# Here we rename identical column names of the binary dummy variables
# These are has_warranty and is_exhangeable
# Code borrowed from: https://datascience.stackexchange.com/questions/41448/how-to-rename-columns-that-have-the-same-name

cols = []
names = ['warranty', 'exchangeable']
count = 0

for column in df_encode_train.columns:
    if column == 'dummy_True':
        cols.append(f'dummy_{names[count]}')
        count+=1
        continue
    cols.append(column)
df_encode_train.columns = cols

cols = []
count = 0

for column in df_encode_test.columns:
    if column == 'dummy_True':
        cols.append(f'dummy_{names[count]}')
        count+=1
        continue
    cols.append(column)
df_encode_test.columns = cols

display(df_encode_train)
display(df_encode_test)
```

## Model Selection

```python
# We can train multiple models on the dataset
# each model will be evaluated using Mean
# Absolute error. This is a regression task, hence
# we need a model suitable for that. 
# Such as Multiple Linear Regression
# or Polynomial Regression. 

# Another possible approach is to make predictions from 
# each model and average them to get the final prediction.

# Model #1: Multiple Linear Regression.
# Model #2: Polynomial Regression.

# Extracting the training/testing data independent and 
# dependent variables

X_train = np.array(df_encode_train.iloc[:, df_encode_train.columns != 'price_usd'])
y_train = np.array(df_encode_train.iloc[:, 3])

X_test = np.array(df_encode_test.iloc[:, df_encode_test.columns != 'price_usd'])
y_test = np.array(df_encode_test.iloc[:, 3])

print(X_test)
print(y_test)

```

### Multiple Linear Regression

```python
# here a Multiple linear regression model is trained
# there aren't any hyper-parameters to tune, hence
# this model is limited. It is also apparent from
# the Exploratory data analysis above that the relationship
# between the independent variables and the dependent variables is
# non-linear. Therefore we need to model this non-linearity.

lin_reg = LinearRegression()

lin_reg.fit(X_train, y_train)
```

### Polynomial Regression

```python
# Each of these models will be trained and evaluated, by increasing the 
# order of the polynomial. However, we do not want to risk overfitting
# the training data. This is why we will steadily increase the order until 
# the mean absolute error stops decreasing or starts increasing.

poly_features_1 = PolynomialFeatures(degree = 2, include_bias = False)
```

```python
# Applying this Polynomial Basis to get the transformed training data

X_train_poly_1 = poly_features_1.fit_transform(X_train)
```

```python
# Applying this Polynomial Basis to get the transformed test data

X_test_poly_1 = poly_features_1.fit_transform(X_test)
```

```python
# fit the model using standard Multiple Linear regression

poly_reg_1 = LinearRegression()

poly_reg_1.fit(X_train_poly_1, y_train)
```

### Evaluation

```python
# Linear Regression prediction

linear_preds = lin_reg.predict(X_test)

# Polynomial Regression prediction

poly_preds_1 = poly_reg_1.predict(X_test_poly_1)

# Hybrid Model Prediction (Average results from each model above)

combined_preds = (linear_preds + poly_preds_1)/2

# calculating mean absolute error

mean_abs_error_1 = abs(linear_preds - y_test)

mean_abs_error_2 = abs(poly_preds_1 - y_test)

mean_abs_error_3 = abs(combined_preds - y_test)


print("Linear: "  + str(mean_abs_error_1.sum(axis=0)/y_test.shape[0]))

print("Poly deg 2: " + str(mean_abs_error_2.sum(axis=0)/y_test.shape[0]))

print("Hybrid model: " + str(mean_abs_error_3.sum(axis=0)/y_test.shape[0]))
```

### Conclusion & Future Work

```python
# As we can see, the polynomial basis regression models the data more accurately
# This is confirmed from the Exploratory Data Analysis performed where it can clearly be seen that
# relationship between the independent variables and target variable (price_usd) is non-linear

# Polynomial regression achieved the lowest mean absolute error and thus is 
# considered the final and best performing model.
# The hybrid model is another approach which also performs well, 
# However this would be more suitable
# for a classification task.

# Future Work:

# Acquire more compuational resources to experiment with higher order polynomials
# and check for overfitting/underfitting. Experiment with other types of models such as SVR
# Random Forrest Regression and MLPs. Also feature scaling might help overcome any bottelnecks 
# in the performance. 
#
# Note, if more RAM is available then more features can be included without affecting running time.
```
